```python
import requests
from datetime import datetime, timedelta

asteriods_data=[]
target = 10000
url= f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-08&api_key={API_KEY}"

start_date = datetime.strptime("2024-01-01", "%Y-%m-%d")
def format_date(date):
    return date.strftime("%Y-%m-%d")
    
while len(asteriods_data) < target:
    end_date = start_date + timedelta(days=7)
    url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={format_date(start_date)}&end_date={format_date(end_date)}&api_key={API_KEY}"

    response = requests.get(url)
    if response.status_code != 200:
        print(f"❌ Error fetching data: {response.status_code}")
        break

    data = response.json()
    near_earth_objects = data.get('near_earth_objects', {})

    for date, asteroids in near_earth_objects.items():
        for ast in asteroids:
            if 'close_approach_data' in ast and ast['close_approach_data']:
                approach = ast['close_approach_data'][0]

                asteriods_data.append({
                    "neo_id": ast['neo_reference_id'],
                    "name": ast['name'],
                    "close_approach_date": approach.get('close_approach_date_full'),
                    "epoch_close_date": approach.get('epoch_date_close_approach'),
                    "velocity_km_s": approach['relative_velocity']['kilometers_per_second'],
                    "miss_distance_km": approach['miss_distance']['kilometers'],
                    "miss_distance_lunar": approach['miss_distance']['lunar'],
                    "orbiting_body": approach['orbiting_body'],
                    "hazardous": ast['is_potentially_hazardous_asteroid']
                })

                if len(asteriods_data) >= target:
                    break          
        if len(asteriods_data) >= target:
            break 
    print(f" Collected: {len(asteriods_data)} entries (up to {format_date(end_date)})")
    start_date = end_date + timedelta(days=1)
print(f"\n Done! Total Neo collected: {len(asteriods_data)}")
```

```python
import requests
from datetime import datetime, timedelta

API_KEY = "your_api_key_here"  # Replace this with your NASA API key

# List to store cleaned asteroid basic info
asteroid_basic_data = []

# Start from Jan 1, 2024
start_date = datetime.strptime("2024-01-01", "%Y-%m-%d")
target = 100  # number of asteroid entries to collect

def format_date(date):
    return date.strftime("%Y-%m-%d")

while len(asteroid_basic_data) < target:
    end_date = start_date + timedelta(days=7)

    url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={format_date(start_date)}&end_date={format_date(end_date)}&api_key={API_KEY}"
    response = requests.get(url)

    if response.status_code != 200:
        print("Error fetching data.")
        break

    data = response.json()
    objects = data.get("near_earth_objects", {})

    for day, asteroids in objects.items():
        for ast in asteroids:
            asteroid = {
                "id": ast.get("id"),
                "name": ast.get("name"),
                "magnitude": float(ast.get("absolute_magnitude_h", 0.0)),
                "dia_min": float(ast.get("estimated_diameter", {}).get("kilometers", {}).get("estimated_diameter_min", 0.0)),
                "dia_max": float(ast.get("estimated_diameter", {}).get("kilometers", {}).get("estimated_diameter_max", 0.0)),
                "hazardous": int(ast.get("is_potentially_hazardous_asteroid", False))
            }
            asteroid_basic_data.append(asteroid)

            if len(asteroid_basic_data) >= target:
                break
        if len(asteroid_basic_data) >= target:
            break

    print(f"✅ Collected: {len(asteroid_basic_data)}")
    start_date = end_date + timedelta(days=1)

print("🎯 Finished collecting basic asteroid data.")
```

```python
import pandas as pd
import streamlit as st
import pymysql

st.title("NASA PROJECT")

# Database connection
myconnection = pymysql.connect(host='127.0.0.1', user='root', passwd='laks',database='NASA')
cur = myconnection.cursor()

# Sidebar query selection
with st.sidebar:
    query_option = st.selectbox(
        "Select a Query",
        [
            "1. Display Entire Table",
            "2. Count of Approaches per Asteroid",
            "3. Average Velocity of Each Asteroid",
            "4. Top 10 Fastest Asteroids",
            "5. Hazardous Asteroids (>3 Approaches)",
            "6. Month with Most Approaches",
            "7. Fastest Ever Approach",
            "8. Sort by Maximum Estimated Diameter",
            "9. Closest Miss Over Time",
            "10. Closest Approach per Asteroid",
            "11. Asteroids > 50,000 km/h",
            "12. Approaches per Month",
            "13. Brightest Asteroid (Lowest Magnitude)",
            "14. Hazardous vs Non-Hazardous Count",
            "15. Asteroids Closer than the Moon or <0.05 AU"
        ],
        placeholder="Choose a query...",
        index=None
    )

# Helpful debug display
st.write("Selected option:", query_option)

# Execute query based on selection
if query_option == "1. Display Entire Table":
    selected_query = "SELECT * FROM asteroids"

elif query_option == "2. Count of Approaches per Asteroid":
    selected_query = "SELECT name, COUNT(*) AS approach_count FROM asteroids GROUP BY name ORDER BY approach_count DESC"

elif query_option == "3. Average Velocity of Each Asteroid":
    selected_query = "SELECT name, AVG(velocity_km_s) AS avg_velocity FROM asteroids GROUP BY name ORDER BY avg_velocity DESC"

elif query_option == "4. Top 10 Fastest Asteroids":
    selected_query = "SELECT name, MAX(velocity_km_s) AS max_velocity FROM asteroids GROUP BY name ORDER BY max_velocity DESC LIMIT 10"

elif query_option == "5. Hazardous Asteroids (>3 Approaches)":
    selected_query = "SELECT name, COUNT(*) AS approach_count FROM asteroids WHERE hazardous = 1 GROUP BY name HAVING approach_count > 3"

elif query_option == "6. Month with Most Approaches":
    selected_query = "SELECT MONTH(close_approach_date) AS month, COUNT(*) AS count FROM asteroids GROUP BY month ORDER BY count DESC LIMIT 1"

elif query_option == "7. Fastest Ever Approach":
    selected_query = "SELECT name, MAX(velocity_km_s) AS fastest_speed FROM asteroids GROUP BY name ORDER BY fastest_speed DESC LIMIT 1"

elif query_option == "8. Sort by Maximum Estimated Diameter":
    selected_query = "SELECT name, MAX(estimated_diameter_max) AS max_diameter FROM asteroids GROUP BY name ORDER BY max_diameter DESC"

elif query_option == "9. Closest Miss Over Time":
    selected_query = """
        SELECT name, close_approach_date, miss_distance_km
        FROM asteroids
        WHERE name = (
            SELECT name FROM asteroids ORDER BY close_approach_date LIMIT 1
        )
        ORDER BY close_approach_date ASC
    """

elif query_option == "10. Closest Approach per Asteroid":
    selected_query = "SELECT name, close_approach_date, MIN(miss_distance_km) AS closest_miss FROM asteroids GROUP BY name"

elif query_option == "11. Asteroids > 50,000 km/h":
    selected_query = "SELECT DISTINCT name FROM asteroids WHERE velocity_km_s > (50000 / 3600)"

elif query_option == "12. Approaches per Month":
    selected_query = "SELECT MONTH(close_approach_date) AS month, COUNT(*) AS total FROM asteroids GROUP BY month ORDER BY month"

elif query_option == "13. Brightest Asteroid (Lowest Magnitude)":
    selected_query = "SELECT name, magnitude FROM asteroids ORDER BY magnitude ASC LIMIT 1"

elif query_option == "14. Hazardous vs Non-Hazardous Count":
    selected_query = "SELECT hazardous, COUNT(DISTINCT name) AS count FROM asteroids GROUP BY hazardous"

elif query_option == "15. Asteroids Closer than the Moon or <0.05 AU":
    selected_query = """
        SELECT name, close_approach_date, miss_distance_km
        FROM asteroids
        WHERE miss_distance_km < (0.05 * 149597870.7) OR miss_distance_lunar < 1
    """

# Run and display the selected query
if query_option:
    cur.execute(selected_query)
    result = cur.fetchall()
    columns = [desc[0] for desc in cur.description]
    df = pd.DataFrame(result, columns=columns)

    st.subheader(query_option)
    st.dataframe(df)
```
